## Q8

```python
r = np.linspace(0.1, 30, 100)
#print(r)
```

```python
menc_halo = MW.MassEnclosed(1, r)
menc_disk = MW.MassEnclosed(2, r)
menc_bulge = MW.MassEnclosed(3, r)
```

```python
Mhalo = 1.975e12 # HW3
```

```python
menc_tot_hern = MW.HernquistMass(r, 60, Mhalo)

fig = plt.figure(figsize = (8, 6))
ax0 = fig.add_subplot(1, 1, 1)
ax0.plot(r, menc_halo, label = 'Halo')
ax0.plot(r, menc_disk, label = 'Disk')
ax0.plot(r, menc_bulge, label = 'Bulge')
ax0.plot(r, menc_halo + menc_disk + menc_bulge, label = 'Total')
ax0.plot(r, menc_tot_hern, label = 'Hernquist')
ax0.semilogy()
ax0.set_xlabel('R [kpc]')
ax0.set_ylabel('Mass [Msun]')
ax0.legend()
plt.show()
```

## Q9

```python
vcirc_halo = MW.CircularVelocity(1, r)
vcirc_disk = MW.CircularVelocity(2, r)
vcirc_bulge = MW.CircularVelocity(3, r)
vcirc_total = MW.CircularVelocityTotal(r)
```

```python
vcirc_hern = MW.HernquistVCirc(r, 60, Mhalo)
```

```python
fig = plt.figure(figsize = (8, 6))
ax0 = fig.add_subplot(1, 1, 1)
ax0.plot(r, vcirc_halo, label = 'Halo')
ax0.plot(r, vcirc_disk, label = 'Disk')
ax0.plot(r, vcirc_bulge, label = 'Bulge')
ax0.plot(r, vcirc_total, label = 'Total')
ax0.plot(r, vcirc_hern, label = 'Hernquist')
ax0.set_xlabel('R [kpc]')
ax0.set_ylabel('Vcirc [km/s]')
ax0.legend()
plt.show()
```
