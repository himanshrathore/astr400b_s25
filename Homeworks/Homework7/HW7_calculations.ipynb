```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
analytic = np.genfromtxt('analytic_orbit.txt')
actual_m33 = np.genfromtxt('Orbit_M33.txt')
actual_m31 = np.genfromtxt('Orbit_M31.txt')
times_analytic = analytic[:, 0]
times_actual = actual_m33[:, 0]
```

```python
def relative(vec1, vec2):
    diff = vec2 - vec1
    mag = np.sqrt(diff[0]**2 + diff[1]**2 + diff[2]**2)
    return mag
```

```python
M31_pos = actual_m31[:, 1:4]
M33_pos = actual_m33[:, 1:4]

M31_vel = actual_m31[:, 4:]
M33_vel = actual_m33[:, 4:]

analytic_M33_pos = analytic[:, 1:4]
analytic_M33_vel = analytic[:, 4:]
```

```python
M33_M31_pos = relative(M33_pos.T, M31_pos.T)
M33_M31_vel = relative(M33_vel.T, M31_vel.T)
a_M33_pos = np.linalg.norm(analytic_M33_pos, axis = 1)
a_M33_vel = np.linalg.norm(analytic_M33_vel, axis = 1)
```

```python
fig = plt.figure(figsize = (8, 6))
ax0 = fig.add_subplot(1, 1, 1)
ax0.plot(times_analytic, a_M33_pos)
ax0.plot(times_actual, M33_M31_pos)
plt.show()
```

```python
fig = plt.figure(figsize = (8, 6))
ax0 = fig.add_subplot(1, 1, 1)
ax0.plot(times_analytic, a_M33_vel)
ax0.plot(times_actual, M33_M31_vel)
plt.show()
```
