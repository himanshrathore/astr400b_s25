```python
import numpy as np
import matplotlib.pyplot as plt
import plots_init as pi
```

```python
MW = np.genfromtxt('./Orbit_MW.txt')
M31 = np.genfromtxt('./Orbit_M31.txt')
M33 = np.genfromtxt('./Orbit_M33.txt')
```

```python
def relative(vec1, vec2):
    diff = vec2 - vec1
    mag = np.sqrt(diff[0]**2 + diff[1]**2 + diff[2]**2)
    return mag
```

```python
times = MW[:, 0]
```

```python
MW_pos = MW[:, 1:4]
M31_pos = M31[:, 1:4]
M33_pos = M33[:, 1:4]

MW_vel = MW[:, 4:]
M31_vel = M31[:, 4:]
M33_vel = M33[:, 4:]
```

```python
M31_MW_pos = relative(M31_pos.T, MW_pos.T)
M33_M31_pos = relative(M33_pos.T, M31_pos.T)
```

```python
fig = plt.figure(figsize = (8, 6))
ax0 = fig.add_subplot(1, 1, 1)
ax0.plot(times, M31_MW_pos)
ax0.semilogy()
ax0.set_title('M31-MW pos')
pi.aesthetic(ax0)
plt.show()
```

```python
fig = plt.figure(figsize = (8, 6))
ax0 = fig.add_subplot(1, 1, 1)
ax0.plot(times, M33_M31_pos)
ax0.set_title('M33-MW pos')
#ax0.set_xlim(6, 10)
#ax0.set_ylim(50, 150)
pi.aesthetic(ax0)
plt.show()
```

```python
M31_MW_vel = relative(M31_vel.T, MW_vel.T)
M33_M31_vel = relative(M33_vel.T, M31_vel.T)
```

```python
fig = plt.figure(figsize = (8, 6))
ax0 = fig.add_subplot(1, 1, 1)
ax0.plot(times, M31_MW_vel)
ax0.set_title('M31-MW vel')
pi.aesthetic(ax0)
plt.show()
```

```python
fig = plt.figure(figsize = (8, 6))
ax0 = fig.add_subplot(1, 1, 1)
ax0.plot(times, M33_M31_vel)
ax0.set_title('M31-M33 vel')
pi.aesthetic(ax0)
plt.show()
```
